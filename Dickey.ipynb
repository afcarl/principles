```python
import pandas as pd
import re
import datetime
import pylab as pl
pl.style.use('fivethirtyeight')
import statsmodels.formula.api as smf
from statsmodels.graphics.api import abline_plot
import statsmodels.api as sma
```

```python
k = pd.read_csv('dickey.csv')
```

```python
k.columns
```

```python
ax = k.plot('Vel', 'AVG', kind='scatter', figsize=(8,8))
mod = smf.ols(formula='AVG ~ Vel', data = k).fit()
print mod.summary()
abline_plot(model_results=mod, ax=ax, color='red', label='Regression')
ax.set_title('Opponent Batting Average vs. R.A. Dickey\'s Knuckleball Speed')
ax.set_xlabel('Average Velocity')
ax.set_ylabel('Batting Average')
```

```python
k['HRPP'] = (k.HR/k.Pitches)*1000
```

```python
ax = k.plot('Vel', 'HRPP', kind='scatter', figsize=(8,8))
mod = smf.ols(formula='HRPP ~ Vel', data = k).fit()
print mod.summary()
abline_plot(model_results=mod, ax=ax, color='red', label='Regression')
ax.set_title('Opponent Batting Average vs. R.A. Dickey\'s Knuckleball Speed')
ax.set_xlabel('Average Velocity')
ax.set_ylabel('Batting Average')
```
